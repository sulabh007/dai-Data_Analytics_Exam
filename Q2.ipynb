```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.drop(columns=['EmployeeCount','Over18'], inplace=True)
```

Histogram of 5 columns

1. Age Range of Employees

```python
sns.histplot(df, x='Age', bins=10, kde=True)
plt.title('Age Range of Employee')

plt.show()
```

2. Monthly Income of the employees

```python
plt.figure(figsize=(10,5))
sns.histplot(df, x='MonthlyIncome', bins=20, kde=True)
plt.title('Monthly Income of Employee')
plt.xticks(range(1000,20001,1000), rotation=90)

plt.show()
```

3. Salary Hike of Employees

```python
sns.histplot(df, x='PercentSalaryHike', bins=14, kde=True)
plt.title('Salary Hike Rate of Employee')
plt.xticks(range(11,26))
plt.show()
```

4. Job Role of Employees

```python
plt.figure(figsize=(10,5))
sns.histplot(df, x='JobRole', bins=10, kde=True)
plt.title('Job Role of Employee')
plt.xticks(rotation=60)
plt.show()
```

5.  Years at company

```python
sns.histplot(df, x='YearsAtCompany', bins=8, kde=True)
plt.title('Years At Company')

plt.show()
```

Pie Charts

1. Pie chart of Married Status of employee

```python
marriedStatus= df['MaritalStatus'].value_counts()
plt.pie(marriedStatus, labels=marriedStatus.index, autopct='%1.1f%%')
```

2. 

```python
education= df['EducationField'].value_counts()
plt.pie(education, labels=education.index, startangle=90, autopct='%1.1f%%', radius=2)
```

```python
dept = df['Department'].value_counts()
plt.pie(dept, labels=dept.index, startangle=90, autopct='%1.1f%%')
plt.title("Department")
```

```python
dept = df['JobRole'].value_counts()
plt.pie(dept, labels=dept.index, startangle=90, autopct='%1.1f%%')
plt.title("Job Role")
```
