```python
df_married= df.where(df['MaritalStatus']=='Married').dropna()
print(df_married.head())
plt.boxplot(df_married)
```

```python
upper = df_married['HourlyRate'].mean() + 3 * df_married['HourlyRate'].std()
lower = df_married['HourlyRate'].mean() - 3 * df_married['HourlyRate'].std()

print(upper)
print(lower)
```

```python
new_df =df_married[(df_married['HourlyRate']<upper) & (df_married['HourlyRate']>lower)]
new_df
```

```python
outliers= df_married[(df_married['HourlyRate']>upper) & (df_married['HourlyRate']<lower)]
outliers
```

There Are No Outliers in Hourly Rate of Married Employees

```python
df_married['Z-score'] = (df_married['HourlyRate'] - df_married['HourlyRate'].mean())/df_married['HourlyRate'].std()
print(df_married.head(10))
```

```python
df_married[df_married['Z-score']>3]
```
